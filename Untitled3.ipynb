```python
def scrape_images_and_text(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Scrape text data
        text_data = ""
        for paragraph in soup.find_all('p'):
            text_data += paragraph.get_text() + '\n\n'

        # Create an HTML file
        html_file = 'output.html'
        with open(html_file, 'w') as f:
            f.write(soup.prettify())

        # Use wkhtmltopdf to generate the PDF
        pdf_file = 'output.pdf'
        try:
            subprocess.run(['wkhtmltopdf', '--encoding', 'utf-8', html_file, pdf_file])
            print("PDF saved successfully as output.pdf")
        except Exception as e:
            print(f"Failed to generate PDF: {e}")
    except Exception as e:
        print(e)
    return text_data
```

```python
start_url = 'https://react.dev/learn'
data = scrape_images_and_text(start_url)
```

```python
chunks = [data[index : index + 512] for index in range(0 , len(data) , 512)]
```
